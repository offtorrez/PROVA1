```python
a = int(input("Digite o primeiro número: "))
b = int(input("Digite o segundo número: "))
c = int(input("Digite o terceiro número: "))
if a <= b and a <= c:
    media = (b + c) / 2
elif b <= a and b <= c:
    media = (a + c) / 2
else:
    media = (a + b) / 2

print("A média dos dois maiores números é:", media)


```
